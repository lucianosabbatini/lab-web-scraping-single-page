```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
```

```python
df_num=df.select_dtypes(include=np.number)
```

```python
X=df_num
```

```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5)
kmeans.fit(X_scaled)
labels = kmeans.predict(X_scaled)
```

```python
# just doing a test before adding to the proper dataframe
df_num['cluster'] = labels
```

```python
df_num
```

```python
# I have created a column with the clusters. but the column goes to end of the dataframe. I want to be the third columns,
# next to the song. So i re-started the kernel and changed to this code to insert the column clusters, through predictions
# on variable labels on the position 2 (third element) through function insert.
df.insert(2, 'cluster', labels)
```

```python
import numpy as np
K = range(2, 20)
inertia = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=1234)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbow Method showing the optimal k')
```
